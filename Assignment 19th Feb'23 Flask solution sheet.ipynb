Q3. What is App routing in Flask? Why do we use app routes?

Answer 3. In Flask, an app route is a URL pattern that the application will match and execute a specific function or view. App routing is a mechanism that allows developers to define URL paths for different parts of their web application, and map those paths to specific Python functions or views.

When a user makes a request to a specific URL, the Flask application searches for a corresponding route that matches the requested URL. Once a matching route is found, the corresponding function or view associated with that route is executed, and the resulting HTML or data is returned to the user.

App routing is important because it helps organize the structure of a Flask application and makes it easy to navigate between different parts of the application. By defining specific routes, developers can create clear and consistent URLs for different views or functions within their application.

Overall, app routing is a fundamental feature of Flask and is critical to building robust and user-friendly web applications.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Answer 5. In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint with the provided method arguments. This function accepts the name of the view function as the first argument and any additional arguments as keyword arguments.

Here's an example code that demonstrates the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is {username}\'s profile'

if __name__ == '__main__':
    with app.test_request_context():
     
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
        print(url_for('user_profile', username='skm_0311'))
        
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
